```python
"""
Welcome to Lesson 12: Working with APIs and Real-World Data — one of the most practical and empowering lessons in your Python mastery journey.

By the end of this lesson, you’ll know how to connect Python to live data sources, pull information from the web, and process it with pandas — just like professional developers and data scientists do every day.


Learning Goals

You’ll learn how to:
Understand what APIs are and how they work
Use the requests library to make HTTP calls
Handle JSON data returned from APIs
Convert API data into pandas DataFrames for analysis
Build a real-world example that pulls live data (e.g., weather, heart health, or exercise data)

What Is an API?

API stands for Application Programming Interface.
It’s a way for two programs to talk to each other — like your Python script talking to a web server.

For example:

Fitbit → share heart rate data

OpenWeatherMap → share weather data

CDC API → share public health data

You send a request → the server sends back a response (usually in JSON format).
"""
```

```python
#Using the requests Library

import requests

response = requests.get("https://api.github.com")
print(response.status_code)
```

```python
# You can also see what came back

print(response.json())
```

```python
"""
Working with JSON Data

Most APIs return data in JSON (JavaScript Object Notation).
It’s structured like Python dictionaries.

Example:
"""

import json

sample = '{"name": "Alex", "age": 35, "city": "Dallas"}'
data = json.loads(sample)
print(data["name"])
```
