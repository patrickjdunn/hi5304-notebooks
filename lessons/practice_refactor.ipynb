## 0) Setup (run once)

This cell:
- finds the repo root (where the `data/` folder lives)
- sets `DATA_DIR` for file loading
- creates a PostgreSQL `engine` for `pd.read_sql(...)`


```python
from pathlib import Path
import os
import pandas as pd
from sqlalchemy import create_engine

# --- Locate repo root (works even if the workspace folder name changes) ---
HERE = Path.cwd()
REPO_ROOT = None
for p in [HERE, *HERE.parents]:
    if (p / "data").exists():
        REPO_ROOT = p
        break

if REPO_ROOT is None:
    raise FileNotFoundError(
        "Could not find a 'data' folder in this workspace. "
        "Make sure your repo contains a top-level data/ directory."
    )

DATA_DIR = REPO_ROOT / "data"
DATA_DIR

# --- Database connection (Codespaces defaults) ---
PGHOST = os.getenv("PGHOST", "db")
PGPORT = os.getenv("PGPORT", "5432")
PGDATABASE = os.getenv("PGDATABASE", "hi5304")
PGUSER = os.getenv("PGUSER", "hi5304_user")
PGPASSWORD = os.getenv("PGPASSWORD", "hi5304")  # course default password

engine = create_engine(
    f"postgresql+psycopg2://{PGUSER}:{PGPASSWORD}@{PGHOST}:{PGPORT}/{PGDATABASE}"
)

SCHEMA = "hi5304"  # course schema name (do not confuse with the GitHub repo name)

def run_sql(query: str) -> pd.DataFrame:
    """Run a SQL query and return a pandas DataFrame."""
    return pd.read_sql(query, engine)

engine
```

```python
schema = SCHEMA
query = f"""
SELECT
    p.patient_id,
    p.first_name,
    p.last_name,
    b.reading_date,
    b.systolic,
    b.diastolic,
    b.heart_rate
FROM {schema}.patients p
JOIN hi5304.bp_readings b
  ON p.patient_id = b.patient_id
ORDER BY b.reading_date;
"""

df_bp = run_sql(query)
df_bp.head()
```

```python
schema = SCHEMA
query = f"""
SELECT
    patient_id,
    COUNT(*) AS medication_count
FROM {schema}.medications
GROUP BY patient_id
ORDER BY medication_count DESC;
"""

df_meds = run_sql(query)
df_meds
```

```python
schema = SCHEMA
query = f"""
SELECT
    AVG(systolic) AS avg_systolic,
    AVG(diastolic) AS avg_diastolic,
    COUNT(*) AS total_readings
FROM {schema}.bp_readings;
"""

df_summary = run_sql(query)
df_summary
```
